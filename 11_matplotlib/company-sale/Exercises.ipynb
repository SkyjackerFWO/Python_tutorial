# Visualizing Company Sale data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/matplotlib/company_sales_data.csv).

```python
csd = pd.read_csv("company_sales_data.csv")
csd.head()
```

### Exercise 1: Read Total profit of all months and show it using a line plot

![img](img/01.png)

```python
csd = csd.sort_values('month_number')

# V biu 
plt.figure(figsize=(10, 6))
plt.plot(csd['month_number'], csd['total_profit'], marker='o')

plt.title('Total Profit Over Months')
plt.xlabel('Month number')
plt.ylabel('Total Profit')

plt.grid(False)
plt.show()
```

### Exercise 2: Get total profit of all months and show line plot with the following Style properties

![img](img/02.png)

```python
import matplotlib.pyplot as plt

# V biu 
fig, ax = plt.subplots()
ax.plot(csd['month_number'], csd['total_profit'],
        marker='o', linestyle='--', color='red', label='Profit data of last year')
plt.legend(loc='lower right')
plt.title('Total Profit Over Months')
plt.xlabel('Month number')
plt.ylabel('Total Profit')

plt.show()
```

### Exercise 3: Read all product sales data and show it  using a multiline plot

![img](img/03.png)

```python
csd.columns
```

```python
# V biu 
fig, ax = plt.subplots()
ax.plot(csd['month_number'], csd['facecream'],
        marker='o', label='Face-cream Sales Data')
ax.plot(csd['month_number'], csd['facewash'],
        marker='o', label='Face-wash Sales Data')
ax.plot(csd['month_number'], csd['toothpaste'],
        marker='o', label='Toothpaste Sales Data')
ax.plot(csd['month_number'], csd['bathingsoap'],
        marker='o', label='Bathingsoap Sales Data')
ax.plot(csd['month_number'], csd['shampoo'],
        marker='o', label='Shampoo Sales Data')
ax.plot(csd['month_number'], csd['moisturizer'],
        marker='o', label='Moisturizer Sales Data')
plt.legend()
plt.title('Sales units in number')
plt.xlabel('Month number')
plt.ylabel('Number of unit Sold')
plt.xticks(csd['month_number'])
plt.show()
```

### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
![img](img/04.png)

```python
fig, ax = plt.subplots()

ax.scatter(csd['month_number'], csd['toothpaste'],
           marker='o', label='Toothpaste Sales Data')

plt.legend()
plt.title('Toothpaste Sales Data')
plt.xlabel('Month number')
plt.ylabel('Number of unit Sold')
plt.grid(True, linestyle="--")
plt.xticks(csd['month_number'])
plt.show()
```

### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

![img](img/05.png)

```python
fix, ax = plt.subplots(figsize=(12,6))

bar_width = 0.25
ax.bar(csd['month_number'], csd['facecream'],
       width=bar_width,
       label='Face-cream Sales Data')
ax.bar(csd['month_number'] + bar_width, csd['facewash'],
       width=bar_width,
       label='Face-wash Sales Data')
plt.legend(loc="upper left")
plt.title('Sales units in number')
plt.xlabel('Month number')
plt.ylabel('Number of unit Sold')
plt.xticks(csd['month_number'])
plt.grid(True)
plt.show()
```

### Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

![img](img/06.png)

```python
fix,ax = plt.subplots()

ax.bar(csd['month_number'], csd['bathingsoap'])

plt.title('Sales units in number')
plt.xlabel('Month number')
plt.ylabel('Number of unit Sold')
plt.xticks(csd['month_number'])
plt.grid(True)
plt.grid(True, linewidth=1, linestyle="--")
plt.savefig("bathingsoapsalesdata.png", dpi=150)
plt.show()
```

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

![img](img/07.png)

```python
profitlist = csd['total_profit'].tolist()
labels = ['low', 'average', 'good', 'best']
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(profitlist, profit_range, label="Profit Data")
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.legend(loc='upper left')
plt.xticks(profit_range)
plt.title('Profit data')
plt.show()
```

### Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

![img](img/08.png)

```python
monthList = csd['month_number'].tolist()
labels = ['FaceCream', 'FaseWash', 'ToothPaste',
          'Bathing soap', 'Shampoo', 'Moisturizer']
salesData = [csd['facecream'].sum(), csd['facewash'].sum(), csd['toothpaste'].sum(),
             csd['bathingsoap'].sum(), csd['shampoo'].sum(), csd['moisturizer'].sum()]

plt.pie(salesData, labels=labels, autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.title('Sales data')
plt.show()
```

### Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

![img](img/09.png)

```python
monthlist = csd['month_number'].tolist()
bathingsoapsalesdata = csd['bathingsoap'].tolist()
facewashsalesdata = csd['facewash'].tolist()

f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(monthlist, bathingsoapsalesdata,
              label="BathingSoap Sales Data", marker="o", color='k', linewidth=3)
axarr[0].set_title("Sales data of Bathingsoap")
axarr[1].plot(monthlist, facewashsalesdata,
              label="Facewash Sales Data", marker="o", color='r', linewidth=3)
axarr[1].set_title('Sales Data of Facewash')
plt.xticks(monthList)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```

### Exercise 10: Read all product sales data and show it using the stack plot

![img](img/10.png)

```python
monthlist = csd['month_number'].tolist()

faceCremSalesData = csd['facecream'].tolist()
faceWashSalesData = csd['facewash'].tolist()
toothPasteSalesData = csd['toothpaste'].tolist()
bathingsoapSalesData = csd['bathingsoap'].tolist()
shampooSalesData = csd['shampoo'].tolist()
moisturizerSalesData = csd['moisturizer'].tolist()

plt.figure(figsize=(13, 6))
plt.plot([], [], color='r', label="Face Cream", linewidth=5)
plt.plot([], [], color='g', label="Face Wash", linewidth=5)
plt.plot([], [], color='b', label="Toothpaste", linewidth=5)
plt.plot([], [], color="c", label="Bathingsoap", linewidth=5)
plt.plot([], [], color="y", label="Shampoo", linewidth=5)
plt.plot([], [], color="m", label="Moisturizer", linewidth=5)

plt.stackplot(monthlist, faceCremSalesData, faceWashSalesData, toothPasteSalesData, bathingsoapSalesData, shampooSalesData,
              moisturizerSalesData, colors=['r', 'g', 'b', 'c', 'm', 'y'])

plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')
plt.title("All Product Sales Data Using Stack Plot")
plt.legend(loc="upper left")
plt.show()
```
